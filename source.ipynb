```python
from PIL import Image, ImageFilter
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame()
for i in range(1,118):
    im = Image.open("data\image ({}).png".format(i))
    im = im.convert('RGB')
    New_image = np.array(im.resize((150, 150)))
    a = pd.DataFrame(New_image.ravel()).T
    df=df.append(a)
```
